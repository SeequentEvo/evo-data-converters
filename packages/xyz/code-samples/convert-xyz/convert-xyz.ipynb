# Import a Pointset Object from an XYZ file

This notebook shows how to load and import a Pointset file from an XYZ file.

To connect with Evo, credentials to login need to be provided, by filling in the environment variables
`CLIENT_ID` and `REDIRECT_URL` and loading them into your Jupyter environment.

Once logged in, a widget will be displayed below allowing you to select an organisation and workspace to publish objects to.

You can then pick from the XYZ files in the `data` subdirectory and choose to upload it.


```python
from evo.notebooks import ServiceManagerWidget

client_id = "your-client-id"
redirect_url = "your-redirect-url"

manager = await ServiceManagerWidget.with_auth_code(
    client_id=client_id,
    redirect_url=redirect_url,
).login()
```

In the cell below we choose the XYZ file we want to publish and set its path in the `xyz_file` variable.

You may also specify tags to add to the new geoscience object.

Then we call `convert_xyz`, passing it the XYZ file path and the service manager widget from above, and finally a path we want the published objects to appear under.


```python
import os
from evo.data_converters.xyz.importer import convert_xyz
from pprint import pp

tags = {"Source": "Jupyter Notebook"}
xyz_file = os.path.join(os.getcwd(), "data/Wolfpass_Lidar_Topo.xyz")
objects_metadata = convert_xyz(
    filepath=xyz_file,
    service_manager_widget=manager,
    tags=tags,
    upload_path="notebook",
    publish_objects=True,
    overwrite_existing_objects=True,
)

print()
print("These objects have now been published:")

for metadata in objects_metadata:
    pp(metadata, indent=4)
```
